## Renombre de variables

Se cambio el nombre de las columnas para que estas fueran mas especificas, bajo las normas de pascal case

```python
df.rename(columns={
    'Código Dane departamento':'codDaneDpto'
    ,'Departamento':'Dpto'
    ,'Código Dane municipio':'codDaneMunicipio'
    ,'Municipio':'Municipio'
    ,'Desagregación cultivo':'desagregacionCultivo'
    ,'Cultivo':'Cultivo'
    ,'Ciclo del cultivo':'cicloDelCultivo'
    ,'Grupo cultivo':'grupoCultivo'
    ,'Subgrupo':'Subgrupo'
    ,'Año':'Anio'
    ,'periodo':'periodoSemestre'
    ,'Área sembrada (ha)':'areaSembradaHa'
    ,'Área cosechada (ha)':'areaCosechadaHa'
    ,'Producción (t)':'produccionTon'
    ,'Rendimiento (t/ha)':'rendimientoTonHa'
    ,'Código del cultivo':'codCultivo'
    ,'Nombre científico del cultivo':'nombreCientificoCultivo'
    ,'Estado físico del cultivo':'estadoFisicoCultivo'
    },
    inplace=True
    )
```

## Datos Nulos

Resvisión general de campos nulos

```python
df.isnull().sum()
```

Al hacer una revision de los datos se encontro que la columna **Cultivo** tiene 66 registros vacios,ya que **desagregacionCultivo** esta relacionada con la columna cultivo allí se visualiza que es una categoria general *Otros*, se decide cambiar este campo vacio de cultivo por el registros del  campo de desagregacionCultivo.

```python
df['Cultivo'] = df['Cultivo'].fillna(df['desagregacionCultivo']) # sustitución
df.isnull().sum()
```

Validación del cambio realizado

```python
df_prueba = df[(df['Cultivo']=='Otros cítricos') | (df['Cultivo']=='Otros caducifolios')]
df_prueba['Cultivo']
```

```python
df.Periodo.value_counts()
```

se realiza un cambio de los registros 'Periodo' por su valor alfabetico correspondiente al semestre del año
* A = Primer semestre
* B = Segundo semestre
* C = año completo 

```python
diccionario_periodo = {
    '2022':'C'
    ,'2021':'C'
    ,'2022A':'A'
    ,'2021A':'A'
    ,'2022B':'B'
    ,'2029A':'A'
    ,'2019A':'A'
    ,'2020':'C'
    ,'2019':'C'
    ,'2021B':'B'
    ,'2020B':'B'
    ,'2019B':'B'
}
```

```python
df[['Anio','Periodo']].head(10)
```

```python
df['Periodo'] = df['Periodo'].map(diccionario_periodo)
```
